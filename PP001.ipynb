```python
# import library
import contractions
import pandas as pd
df = pd.read_csv("SentimentTrain.csv") #loaded the dataframe

def expand_contractions(text):
    expanded_words = [contractions.fix(word) for word in text.split()]
    expanded_text = ' '.join(expanded_words)
    return expanded_text

# Apply the function to the 'text' column
df['expanded_text'] = df['text'].apply(expand_contractions)

# Display the dataframe with the expanded text
print(df[['text', 'expanded_text', 'label']])

#saving the updated dataframe to a new CSV file
df.to_csv("expdata.csv", index=False)
```
